```python
#Glucose
df["Glucose"] = df["Glucose"].replace({0:np.nan})
df["Glucose"].isnull().sum()
df["Glucose"] =df["Glucose"].fillna(df["Glucose"].mean())
df["Glucose"].isnull().sum()
```

```python
#BloodPressure
df["BloodPressure"] = df["BloodPressure"].replace({0:np.nan})
df["BloodPressure"].isnull().sum()
df["BloodPressure"] =df["BloodPressure"].fillna(df["BloodPressure"].mean())
df["BloodPressure"].isnull().sum()
```

```python
#SkinThickness
df["SkinThickness"] = df["SkinThickness"].replace({0:np.nan})
df["SkinThickness"].isnull().sum()
df["SkinThickness"] =df["SkinThickness"].fillna(df["SkinThickness"].mean())
df["SkinThickness"].isnull().sum()
```

```python
#Insulin
df["Insulin"] = df["Insulin"].replace({0:np.nan})
df["Insulin"].isnull().sum()
df["Insulin"] =df["Insulin"].fillna(df["Insulin"].mean())
df["Insulin"].isnull().sum()
```

```python
#BMI
df["BMI"] = df["BMI"].replace({0:np.nan})
df["BMI"].isnull().sum()
df["BMI"] =df["BMI"].fillna(df["BMI"].mean())
df["BMI"].isnull().sum()
```

```python
x = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
#loading the libaries 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
#spliting the test and train data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=34)
```

```python
#initializing the models
dt = DecisionTreeClassifier()
knn = KNeighborsClassifier()
svm = SVC()
rf = RandomForestClassifier()
```

```python
#training the models
model_dt = dt.fit(x_train,y_train)
model_rf = rf.fit(x_train,y_train)
model_svm = svm.fit(x_train,y_train)
model_knn = knn.fit(x_train,y_train)
```

```python
#loading validaiton packcages 
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_curve
```

```python
#predicitons
dt_pred = model_dt.predict(x_test)
rf_pred = model_rf.predict(x_test)
svm_pred = model_svm.predict(x_test)
knn_pred = model_knn.predict(x_test)
```

```python
acc_dt = accuracy_score(y_test,dt_pred)
acc_rf = accuracy_score(y_test,rf_pred)
acc_svm = accuracy_score(y_test,svm_pred)
acc_knn = accuracy_score(y_test,knn_pred)
```

```python
print("DecisionTreeClassifier accuracy :",acc_dt*100,"percent.\n")
print("RandomForestClassifier accuracy :",acc_rf*100,"percent.\n")
print("SVM  accuracy :",acc_svm*100,"percent.\n")
print("KNN accuracy :",acc_knn*100,"percent.\n")
```

```python
#SVM performs well among others.
```
